```python
df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace=True)
df['Alley'].fillna('None', inplace=True)
df['MasVnrType'].fillna('None', inplace=True)
df['MasVnrArea'].fillna(0, inplace=True)
df['BsmtQual'].fillna('None', inplace=True)
df['BsmtCond'].fillna('None', inplace=True)

bsmt = df[['Id', 'BsmtQual', 'BsmtExposure']]
bsmt_exposure_to_fill = bsmt[(bsmt['BsmtExposure'].isna() == True) & (bsmt['BsmtQual'] != 'None')]['Id'].values[0]
df[df['Id'] == bsmt_exposure_to_fill]['BsmtExposure'].fillna('No', inplace=True)
df['BsmtExposure'].fillna('None', inplace=True)

df['BsmtFinType1'].fillna('None', inplace=True)

bsmt = df[['Id', 'BsmtQual', 'BsmtFinType2']]
bsmt_fintype2_to_fill = bsmt[(bsmt['BsmtFinType2'].isna() == True) & (bsmt['BsmtQual'] != 'None')]['Id'].values[0]
df[df['Id'] == bsmt_fintype2_to_fill]['BsmtFinType2'].fillna('Unf', inplace=True)
df['BsmtFinType2'].fillna('None', inplace=True)

df['Electrical'].fillna('Mix', inplace=True)
df['FireplaceQu'].fillna('None', inplace=True)
df['GarageType'].fillna('None', inplace=True)
df['GarageYrBlt'].fillna(0, inplace=True)
df['GarageFinish'].fillna('None', inplace=True)
df['GarageQual'].fillna('None', inplace=True)
df['GarageCond'].fillna('None', inplace=True)
df['PoolQC'].fillna('None', inplace=True)
df['Fence'].fillna('None', inplace=True)
df['MiscFeature'].fillna('None', inplace=True)
```

```python
df.drop('Id', axis='columns', inplace=True)
df.describe()
```

```python
print(f'Before removing outliers: {df.shape}')
outliers = pd.DataFrame()
numeric_cols = df.describe().columns
string_cols = df.drop(numeric_cols, axis='columns')

for col, row in df[numeric_cols].items():
    upper_limit = df[col].mean() + 3 * df[col].std()
    lower_limit = df[col].mean() - 3 * df[col].std()

    outliers = pd.concat([outliers, df[(df[col] < lower_limit) | (df[col] > upper_limit)]])

outliers.drop_duplicates(inplace=True)
df.drop(outliers.index, axis='index', inplace=True)
print(f'After removing outliers: {df.shape}')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
```

```python
X = df.drop('SalePrice', axis='columns')
y = df['SalePrice']

le = LabelEncoder()
for col, row in X.items():
    if col in string_cols:
        X[col] = le.fit_transform(X[col])

scaler = MinMaxScaler()
X = scaler.fit_transform(X, y)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
models = {
    'Linear': {
        'model': LinearRegression(),
        'params': {}
    },
    'Lasso': {
        'model': Lasso(),
        'params': {
            'alpha': [110]
        }
    },
    'Ridge': {
        'model': Ridge(),
        'params': {
            'alpha': [2]
        }
    },
    'SVM': {
        'model': SVR(),
        'params': {
            'C': [17443],
            'kernel': ['poly']
        }
    }
}
```

There isn't a lot of parameters but that's because I tested so many, so the ones that performed best multiple times I just stuck with so the testing would go faster and this is what came out at the end.

```python
scores = []
best_estimators = {}

for mn, mp in models.items():
    gscv = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    gscv.fit(X_train, y_train)

    scores.append({
        'Model': mn,
        'Best Params': gscv.best_params_,
        'Best Score': gscv.best_score_
    })
    best_estimators[mn] = gscv.best_estimator_

pd.DataFrame(scores, columns=['Model', 'Best Params', 'Best Score'])
```

```python
best_estimators
```

```python
best_estimators['Linear'].score(X_test, y_test), best_estimators['Lasso'].score(X_test, y_test), best_estimators['Ridge'].score(X_test, y_test), best_estimators['SVM'].score(X_test, y_test)
```

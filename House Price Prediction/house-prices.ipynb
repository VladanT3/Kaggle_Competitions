```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.shape
```

```python
nulls = df.isna().sum()
nulls[:10], nulls[10:20], nulls[20:30], nulls[30:40], nulls[40:50], nulls[50:60], nulls[60:70], nulls[70:80], nulls[80:]
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace=True)
df['Alley'].fillna('None', inplace=True)
df['MasVnrType'].fillna('None', inplace=True)
df['MasVnrArea'].fillna(0, inplace=True)
df['BsmtQual'].fillna('None', inplace=True)
df['BsmtCond'].fillna('None', inplace=True)

bsmt = df[['Id', 'BsmtQual', 'BsmtExposure']]
bsmt_exposure_to_fill = bsmt[(bsmt['BsmtExposure'].isna() == True) & (bsmt['BsmtQual'] != 'None')]['Id'].values[0]
df[df['Id'] == bsmt_exposure_to_fill]['BsmtExposure'].fillna('No', inplace=True)
df['BsmtExposure'].fillna('None', inplace=True)

df['BsmtFinType1'].fillna('None', inplace=True)

bsmt = df[['Id', 'BsmtQual', 'BsmtFinType2']]
bsmt_fintype2_to_fill = bsmt[(bsmt['BsmtFinType2'].isna() == True) & (bsmt['BsmtQual'] != 'None')]['Id'].values[0]
df[df['Id'] == bsmt_fintype2_to_fill]['BsmtFinType2'].fillna('Unf', inplace=True)
df['BsmtFinType2'].fillna('None', inplace=True)

df['Electrical'].fillna('Mix', inplace=True)
df['FireplaceQu'].fillna('None', inplace=True)
df['GarageType'].fillna('None', inplace=True)
df['GarageYrBlt'].fillna(0, inplace=True)
df['GarageFinish'].fillna('None', inplace=True)
df['GarageQual'].fillna('None', inplace=True)
df['GarageCond'].fillna('None', inplace=True)
df['PoolQC'].fillna('None', inplace=True)
df['Fence'].fillna('None', inplace=True)
df['MiscFeature'].fillna('None', inplace=True)
```
